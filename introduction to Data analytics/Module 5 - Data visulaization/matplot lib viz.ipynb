```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# create dataset
height = [3, 12, 5, 18, 45]
bars = ("A", "B", "C", "D", "E")

df = pd.DataFrame({"Category": bars, "Height": height})
```

```python
# Create bars
plt.bar(x=df["Category"], height=df["Height"])

# Create names on the x-axis
plt.xticks(ticks=df["Category"], labels=df["Height"])

# Show graphic
plt.show()
```

```python
# Create bars
df = df.sort_values(by=["Height"])

plt.bar(x=df["Category"], height=df["Height"])

# Create names on the x-axis
plt.xticks(ticks=df["Category"], labels=df["Height"])

# Show graphic
plt.show()
```

```python
# Create horizontal bars
plt.barh(y=df["Category"], width=df["Height"])

# Show graphic
plt.show()
```

```python
# Sort the table
df = df.sort_Heights(by=['Height'])

# Create horizontal bars
plt.barh(y=df.Category, width=df.Height)

# Display the plot
plt.show()
```

```python
# library
import matplotlib.pyplot as plt

# Create bars
store1_sales = [3, 3, 1]
store2_sales = [4, 2, 3]
store3_sales = [4, 6, 7, 10, 4, 4]
total_sales = store1_sales + store2_sales + store3_sales

# The X position of bars
store1_pos = [1, 5, 9]
store2_pos = [2, 6, 10]
store3_pos = [3, 4, 7, 8, 11, 12]
all_pos = store1_pos + store2_pos + store3_pos


barWidth = 0.9
# Create barplot
plt.bar(
    store1_pos,
    store1_sales,
    width=barWidth,
    color=(0.3, 0.1, 0.4, 0.6),
    label="Store 1",
)
plt.bar(
    store2_pos,
    store2_sales,
    width=barWidth,
    color=(0.3, 0.5, 0.4, 0.6),
    label="Store 2",
)
plt.bar(
    store3_pos,
    store3_sales,
    width=barWidth,
    color=(0.3, 0.9, 0.4, 0.6),
    label="Store 3",
)

# Create legend
plt.legend()

# Text below each barplot with a rotation at 90°
plt.xticks(
    [r + barWidth for r in range(len(all_pos))],
    [
        "Electronics (S1)",
        "Furniture (S1)",
        "Clothing (S1)",
        "Toys (S1)",
        "Electronics (S2)",
        "Furniture (S2)",
        "Clothing (S2)",
        "Toys (S2)",
        "Electronics (S3)",
        "Furniture (S3)",
        "Clothing (S3)",
        "Toys (S3)",
    ],
    rotation=90,
)

# Create labels
labels = [
     "6k",
     "25k",
     "13k",
     "36k",
     "30k",
     "11k",
     "16k",
     "37k",
     "14k",
     "4k",
     "31k",
     "34k",
]

# Text on the top of each bar
for i in range(len(all_pos)):
    plt.text(x=all_pos[i]- barWidth/4 , y=total_sales[i] + 0.1, s=labels[i], size=6)

# Adjust the margins
plt.subplots_adjust(bottom=0.2, top=0.98)

# Show graphic
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
store1_sales = [3, 3, 1]
store2_sales = [4, 2, 3]
store3_sales = [4, 6, 7, 10, 4, 4]

store1_categories = ["Electronics", "Furniture", "Clothing"]
store2_categories = ["Electronics", "Furniture", "Clothing"]
store3_categories = [
    "Electronics",
    "Furniture",
    "Clothing",
    "Toys",
    "Electronics",
    "Furniture",
]

store1_names = ["Store 1"] * len(store1_sales)
store2_names = ["Store 2"] * len(store2_sales)
store3_names = ["Store 3"] * len(store3_sales)

# Combine the data
sales_data = {
    "store_name": store1_names + store2_names + store3_names,
    "product_category": store1_categories + store2_categories + store3_categories,
    "sales": store1_sales + store2_sales + store3_sales,
}

df = pd.DataFrame(sales_data)

# Create bar plot using DataFrame
barWidth = 0.9
plt.figure(figsize=(10, 6))

# Unique categories and store names
categories = df["product_category"].unique()
store_names = df["store_name"].unique()

# Define colors
colors = [(0.3, 0.1, 0.4, 0.6), (0.3, 0.5, 0.4, 0.6), (0.3, 0.9, 0.4, 0.6)]

# Plot each store's sales
for i, store in enumerate(store_names):
    store_data = df[df["store_name"] == store]
    store_pos = list(range(i, len(store_data) * len(store_names), len(store_names)))
    plt.bar(
        store_pos, store_data["sales"], width=barWidth, color=colors[i], label=store
    )

# Create legend
plt.legend()

# Text below each barplot with a rotation at 90°
plt.xticks(
    [r + barWidth for r in range(len(df))],
    [
        f"{cat} ({store})"
        for cat, store in zip(df["product_category"], df["store_name"])
    ],
    rotation=90,
)

# Create labels
labels = [
    "6k",
    "25k",
    "13k",
    "36k",
    "30k",
    "11k",
    "16k",
    "37k",
    "14k",
    "4k",
    "31k",
    "34k",
]

# Text on the top of each bar
for i in range(len(df)):
    plt.text(x=i - barWidth / 4, y=df["sales"].iloc[i] + 0.1, s=labels[i], size=6)

# Adjust the margins
plt.subplots_adjust(bottom=0.2, top=0.98)

# Show graphic
plt.show()
```
