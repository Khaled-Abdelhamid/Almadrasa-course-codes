# seaborn

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
df
```

# Plotting simple violin plot

```python
# plot
sns.violinplot(x=df["sepal_length"])
```

plotting to see two species

```python
sns.violinplot(data=df[["sepal_length", "petal_length"]])
plt.show()
```

# PLotting all species

```python
# plot
sns.violinplot(y=df["species"], x=df["sepal_length"])
```

# change line width

```python
# Change line width
sns.violinplot(x=df["species"], y=df["sepal_length"], linewidth=2)
plt.show()
```

# Change color


```python
# plot
sns.violinplot(x=df["species"], y=df["sepal_length"], color="red")
plt.show()
```

# Using a color palette


```python
# Use a color palette

sns.violinplot(hue=df["species"],x=df["species"], y=df["sepal_length"], palette="Blues",legend=True)

plt.show()
```

## ordering violin plots

```python
# specifying the group list as 'order' parameter and plotting
sns.violinplot(
    x="species", y="sepal_length", data=df, order=["versicolor", "virginica", "setosa"]
)
plt.show()
```

# Grouped violin plot

```python
df = sns.load_dataset("tips")
df
```

```python
# Grouped violinplot
sns.violinplot(x="day", y="total_bill", hue="smoker", data=df, palette="Pastel1")
plt.show()
```

```python
# Grouped violinplot
sns.violinplot(x="smoker", y="total_bill", hue="day", data=df, palette="Pastel1")
plt.show()
```
