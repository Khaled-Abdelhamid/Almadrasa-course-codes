 # Matplotlib violin plots

```python
# Libraries & dataset
import matplotlib.pyplot as plt
import pandas as pd
```

# Load dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
df
```

# Simple violin plot

```python
plt.violinplot(df["sepal_length"], vert=False)
plt.xlabel("Sepal Length")
plt.show()
```

# Plotting to see two species

```python
# Violin plot for sepal_length and sepal_width
fig, ax = plt.subplots()
data = [df["sepal_length"], df["sepal_width"]]
ax.violinplot(data)
ax.set_xticks([1, 2])
ax.set_xticklabels(["Sepal Length", "Sepal Width"])
plt.show()
```

# Plotting all species

```python
# Violin plot for sepal_length by species
species = df["species"].unique()
data = [df[df["species"] == s]["sepal_length"] for s in species]
(type(data[0]))
```

```python
fig, ax = plt.subplots()
ax.violinplot(data)
ax.set_xticks([1,2,3])
ax.set_xticklabels(species)
ax.set_ylabel("Sepal Length")
plt.show()
```

# Customize violin plot

 Uniform color

```python
# Uniform color
fig, ax = plt.subplots()
parts = ax.violinplot(data, widths=0.4)
for pc in parts["bodies"]:
    pc.set_facecolor("red")
    pc.set_edgecolor("yellow")
    
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(species)
ax.set_ylabel("Sepal Length")
plt.show()
```

 ## Ordering violin plots

 Directly specifying a list

```python
# Specifying the group list as 'order' parameter and plotting
order = ["versicolor", "virginica", "setosa"]
data = [df[df["species"] == s]["sepal_length"] for s in order]

fig, ax = plt.subplots()
ax.violinplot(data)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(order)
ax.set_ylabel("Sepal Length")
plt.show()
```
