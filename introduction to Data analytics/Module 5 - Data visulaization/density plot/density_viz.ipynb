```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
df
```

Simple Density plot

```python
# Make default density plot
sns.kdeplot(df["sepal_width"])
```

Density plot with fill

```python
sns.kdeplot(df["sepal_width"], fill=True)
plt.show()
```

change color in density plot

horizontal density plot

```python
sns.kdeplot(y=df["sepal_width"], fill=True, color="skyblue")
plt.show()
```

Density plot with choosing the smoothing bandwidth to use

```python
sns.kdeplot(df["sepal_width"], fill=True, color="olive", bw_method=0.1)
plt.show()
```

```python
sns.kdeplot(df["sepal_width"], fill=True, color="olive", bw_method=0.1)
plt.show()
```

```python
sns.kdeplot(df["sepal_width"], fill=True, color="olive", bw_method=0.2)
plt.show()
```

Multiple density plots at once

```python
sns.kdeplot(df["sepal_width"], fill=True)
sns.kdeplot(df["sepal_length"], fill=True)
plt.show()
```

```python
# Set seaborn whitegrid theme
sns.set_theme(style="whitegrid")

# Without transparency
sns.kdeplot(
    data=df, x="sepal_width", hue="species", fill=True, alpha=0.3)
plt.show()
```

```python

tips = sns.load_dataset("tips")

# Create a grid
g = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)

# Draw density plots
g = g.map(
    sns.kdeplot,
    "total_bill",
    fill=True,
    alpha=1,
    legend=False,
)

# Show the graph
plt.show()
```
