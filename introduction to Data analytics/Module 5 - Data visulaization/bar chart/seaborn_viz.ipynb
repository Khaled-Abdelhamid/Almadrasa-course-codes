```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# Bar plots

```python
# create dataset
height = [3, 12, 5, 18, 45]
bars = ("A", "B", "C", "D", "E")

df = pd.DataFrame({"Category": bars, "Height": height})
df
```

```python
# Assuming df is already defined and contains columns "Category" and "Height"
sns.barplot(x=df["Category"], y=df["Height"])
```

```python

# Assuming df is already defined and contains columns "Category" and "Height"
sns.barplot(x="Category", y="Height", data=df)
```

```python
sorted_categories = df.sort_values(by="Height",ascending=True)
sorted_categories
```

```python
sns.barplot(x="Category", y="Height", data=df, order=sorted_categories["Category"])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the color with RGBA
color = (
    0.2,  # redness
    0.4,  # greenness
    0.2,  # blueness
    0.6,  # transparency
)

# Create a seaborn barplot
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x="Category", y="Height", data=df, ax=ax, color=color)

# Show graph
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the colors for each category
color = ["lightblue", "blue", "purple", "red", "black"]

# Create a seaborn barplot
fig, ax = plt.subplots()
sns.barplot(x="Category", y="Height", data=df ,hue="Category", palette=color)

# Show graph
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the colors and edge colors
colors = ["lightblue", "blue", "purple", "red", "black"]
edgecolors = ["red", "orange", "darkblue", "darkred", "yellow"]

# Create a seaborn barplot with base colors
fig, ax = plt.subplots()
sns.barplot(x="Category", y="Height", data=df, palette=colors, ax=ax)

# Loop through the bars to set edge colors
for bar, edgecolor in zip(ax.patches, edgecolors):
    bar.set_edgecolor(edgecolor)
    bar.set_linewidth(3)
    bar.set_alpha(0.8)

# Show graph
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the color with RGBA
color = (0.5, 0.1, 0.5, 0.6)

# Create a seaborn barplot
fig, ax = plt.subplots()
sns.barplot(x="Category", y="Height", data=df, color=color, ax=ax)

# Add title and axis labels
ax.set_title("My title")
ax.set_xlabel("categories")
ax.set_ylabel("values")

# Show graph
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the color with RGBA
color = (0.5, 0.1, 0.5, 0.6)

# Create a seaborn barplot
fig, ax = plt.subplots()
sns.barplot(x="Category", y="Height", data=df, color=color, ax=ax)

# Add title and axis labels with custom formatting
ax.set_title("My title", fontweight="bold", color="orange", fontsize=18)
ax.set_xlabel("category", fontweight="bold", color="orange", fontsize=18)
ax.set_ylabel("values", fontweight="bold", color="orange", fontsize=18)

# Show graph
plt.show()
```

## Grouped plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    "Store": ["Store 1"] * 3 + ["Store 2"] * 3 + ["Store 3"] * 6,
    "Category": [
        "Electronics",
        "Furniture",
        "Clothing",
        "Electronics",
        "Furniture",
        "Clothing",
        "Electronics",
        "Furniture",
        "Clothing",
        "Toys",
        "Appliances",
        "Books",
    ],
    "Sales": [3000, 3000, 1000, 4000, 2000, 3000, 4000, 6000, 7000, 10000, 4000, 4000],
}

df = pd.DataFrame(data)
df
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Create barplot
barplot = sns.barplot(data=df, x="Category", y="Sales", hue="Store")

# Grid
plt.grid(axis="y", alpha=0.5)

# Show graphic
plt.show()
```

## Stacked / grouped Bar plots

```python
# Import necessary libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Data
sales_data = {
    "Period": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Smartphone": [120, 300, 150, 180, 220],
    "Tablet": [280, 60, 160, 50, 100],
    "Laptop": [290, 30, 240, 250, 170],
}

# Convert data to a pandas DataFrame
df = pd.DataFrame(sales_data)
df
# Melt the DataFrame to long-form for seaborn
df_melted = df.melt(id_vars="Period", var_name="Device", value_name="Sales")
df_melted
```

```python
# Melt the DataFrame to long-form for seaborn
df_melted = df.melt(id_vars="Period", var_name="Device", value_name="Sales")
df_melted
```

```python
# Create the seaborn barplot
plt.figure(figsize=(10, 6))
sns.barplot(x="Device", y="Sales", hue="Period", data=df_melted)

# Add labels and title
plt.xlabel("Period", fontweight="bold")
plt.ylabel("Sales Figures", fontweight="bold")
plt.title(
    "Monthly Sales Figures for Different Consumer Electronics",
    fontsize=13,
    fontweight="bold",
)

# Show the plot
plt.show()
```
