```python
# <class 'pandas.core.frame.DataFrame'>
# RangeIndex: 777 entries, 0 to 776
# Data columns (total 12 columns):
#  #   Column      Non-Null Count  Dtype
# ---  ------      --------------  -----
#  0   Country     777 non-null    object
#  1   Rank        777 non-null    int64
#  2   Score       777 non-null    float64
#  3   Support     777 non-null    float64
#  4   GDP         777 non-null    float64
#  5   Health      777 non-null    float64
#  6   Freedom     777 non-null    float64
#  7   Generosity  777 non-null    float64
#  8   Corruption  777 non-null    float64
#  9   Year        777 non-null    int64
#  10  target      777 non-null    object
#  11  target_n    777 non-null    int64
# dtypes: float64(7), int64(3), object(2)
# memory usage: 73.0+ KB


# Function to plot all pandas DataFrame plots with customizations
def plot_dataframe(df):
    # Line plot
    df.plot(title="Line Plot", color=["r", "g", "b", "c", "m", "y", "k"])
    plt.show()

    # Bar plot
    df.plot.bar(title="Bar Plot", color=["r", "g", "b", "c", "m", "y", "k"])
    plt.show()

    # Horizontal bar plot
    df.plot.barh(title="Horizontal Bar Plot", color=["r", "g", "b", "c", "m", "y", "k"])
    plt.show()

    # Histogram
    df.plot.hist(title="Histogram", bins=15, alpha=0.7)
    plt.show()

    # Box plot
    df.plot.box(title="Box Plot")
    plt.show()

    # KDE (Kernel Density Estimate) plot
    df.plot.kde(title="KDE Plot")
    plt.show()

    # Area plot
    df.plot.area(title="Area Plot", alpha=0.4)
    plt.show()

    # Scatter plot
    df.plot.scatter(x="GDP", y="Score", title="Scatter Plot: GDP vs Score", color="r")
    plt.show()

    # Grouped bar plot using 'by' keyword
    df.groupby("target").mean().plot.bar(
        title="Grouped Bar Plot by Target", color=["r", "g", "b", "c", "m", "y", "k"]
    )
    plt.show()

    # Grouped horizontal bar plot using 'by' keyword
    df.groupby("target").mean().plot.barh(
        title="Grouped Horizontal Bar Plot by Target",
        color=["r", "g", "b", "c", "m", "y", "k"],
    )
    plt.show()


# Function to plot all pandas Series plots with customizations
def plot_series(series):
    # Line plot
    series.plot(title="Line Plot", color="r")
    plt.show()

    # Bar plot
    series.plot.bar(title="Bar Plot", color="r")
    plt.show()

    # Horizontal bar plot
    series.plot.barh(title="Horizontal Bar Plot", color="r")
    plt.show()

    # Histogram
    series.plot.hist(title="Histogram", bins=15, alpha=0.7, color="r")
    plt.show()

    # Box plot
    series.plot.box(title="Box Plot")
    plt.show()

    # KDE (Kernel Density Estimate) plot
    series.plot.kde(title="KDE Plot")
    plt.show()

    # Area plot
    series.plot.area(title="Area Plot", alpha=0.4)
    plt.show()

    # Pie plot
    series.plot.pie(title="Pie Plot", autopct="%1.1f%%")
    plt.show()


# Plot all DataFrame plots
plot_dataframe(df)

# Plot all Series plots (example using 'Score' column)
plot_series(df["Score"])
```
