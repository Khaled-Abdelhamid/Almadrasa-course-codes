```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
df
```

# Simple histogram

```python
plt.hist(df["sepal_length"])
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

# Customizing number of bins

```python
plt.hist(df["sepal_length"], bins=20)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

# Customizing Histogram appearance

```python
# 'bins=20': Divides the range of x values (sepal lengths) into 20 equal intervals
# 'color="blue"': Sets the color of the bars in the histogram to blue
# 'edgecolor="black"': Sets the color of the bar edges to black
# 'rwidth=0.5': Specifies the relative width of the bars as a fraction of the bin width

plt.hist(df["sepal_length"], bins=20, color="blue", edgecolor="black",linewidth=3, rwidth=1)
plt.axvline(x=df["sepal_length"].mean(), color="black", linestyle=":")

plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

# Plot Histogram for each categorical value in separate figure

```python
# Plot the histograms of each group
for group in df["species"].unique():
    filtered_df = df[df["species"] == group]
    plt.hist(filtered_df["sepal_length"], bins=10)
    plt.title(f"Histogram of Sepal Length for {group}")
    plt.xlabel("Sepal Length")
    plt.ylabel("Frequency")
    plt.show()
```

# Multi histogram in one figure

```python
# Plot histograms with different colors for each species
group_name = df["species"].unique()
colors = ["purple", "orange", "green"]
for i, group in enumerate(group_name):
    plt.hist(
        df[df["species"] == group]["sepal_length"],
        bins=12,
        edgecolor="gray",
        color=colors[i],
        alpha=0.5,
        label=group,
    )

plt.legend()
plt.title("Distribution of Sepal Length for Different Species")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```
