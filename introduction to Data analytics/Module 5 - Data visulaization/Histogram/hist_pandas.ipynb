```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
df
```

# Simple histogram

```python
# Plot histograms for 'sepal_length'
# df["sepal_length"].hist()
# df["sepal_length"].plot(kind="hist")
df["sepal_length"].plot.hist()

plt.show()
```

```python
# Plot histograms with additional customization
ax = df["sepal_length"].hist()
plt.title("histogram of sepal length")
plt.show()
```

# Customizing Histogram appearance

```python
ax = df["sepal_length"].hist(
    grid=False,
    xlabelsize=10,
    ylabelsize=12,
    bins=20,
    edgecolor="black",
    color="orange",
    rwidth=0.8,
)
ax.axvline(x=df["sepal_length"].mean(), color="black", linestyle="--")
ax.set_title("Distribution of Sepal Length", weight="bold")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Frequency")
plt.show()
```

# Plot Histogram for each categorical value in separate figure

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and it's already defined
for group in df["species"].unique():
    filtered_df = df[df["species"] == group]

    # Use Pandas to filter the DataFrame based on species and then plot
    filtered_df["sepal_length"].hist(
        bins=10, label=f"Histogram of Sepal Length for {group}"
    )

    plt.title(f"Histogram of Sepal Length for {group}")
    plt.xlabel("Sepal Length")
    plt.ylabel("Frequency")
    plt.show()
```

# Multi histogram in one figure

```python
# Plot histograms with different colors for each species
group_name = df["species"].unique()
colors = ["purple", "orange", "green"]

for i, group in enumerate(group_name):
    filtered_df = df[df["species"] == group]
    ax = filtered_df["sepal_length"].hist(
        figsize=(8, 4), edgecolor="gray", bins=12, color=colors[i],alpha=0.5
    )

ax.legend(group_name)
ax.set_title("Distribution of Sepal Length for Different Species")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Frequency")
plt.show()
```
