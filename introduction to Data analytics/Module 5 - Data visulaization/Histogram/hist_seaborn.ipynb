```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
```

```python
df = sns.load_dataset('iris')
df
```

# Simple Histogram

```python
sns.histplot(df["sepal_length"])
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

# Changing number of bins

```python
sns.histplot(df["sepal_length"], bins=20)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

# Customizing Histogram appearance

```python
ax =sns.histplot(
    df["sepal_length"],
    bins=20,
    color="blue",
    edgecolor="black")

for patch in ax.patches:
    patch.set_width(patch.get_width() * 1)
    

plt.axvline(x=df["sepal_length"].mean(), color="black", linestyle="--")
plt.title("Distribution of Sepal Length", weight="bold")
plt.xlabel("Sepal Length", fontsize=10)
plt.ylabel("Frequency", fontsize=12)
plt.grid(False)
plt.show()
```

# Plot Histogram for each categorical value in separate figure

```python
# Plot the histograms of each group
for group in df["species"].unique():
    filtered_df = df[df["species"] == group]
    sns.histplot(filtered_df["sepal_length"], bins=10)
    plt.title(f"Histogram of Sepal Length for {group}")
    plt.xlabel("Sepal Length")
    plt.ylabel("Frequency")
    plt.show()
```

# Adding box plot

```python
# creating a figure composed of two matplotlib.Axes objects (ax1 and ax2)
f, (ax1, ax2) = plt.subplots(
    2, sharex=True, gridspec_kw={"height_ratios": (0.15, 0.85)}
)

# assigning a graph to each ax
sns.boxplot(data=df["sepal_length"], orient="h", ax=ax1)
sns.histplot(data=df, x="sepal_length", ax=ax2)

plt.show()
```

# Multi histogram in one axes

```python
# Plot histograms with different colors for each species
sns.histplot(
    data=df,
    x="sepal_length",
    hue="species",
    bins=12,
    edgecolor="gray",
    alpha=0.7,
    legend=True
)
plt.title(label="Distribution of Sepal Length for Different Species")
plt.xlabel(xlabel="Sepal Length")
plt.ylabel(ylabel="Frequency")
plt.show()
```

## Stacking of histogram

```python
sns.histplot(
    df,
    x="sepal_length",
    hue="species",
    bins=12,
    edgecolor="gray",
    alpha=0.7,
    multiple="stack",
)

plt.title("Distribution of Sepal Length for Different Species")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

## Changing element of histogram

```python
# Plot histograms with different colors for each species
sns.histplot(
    df,
    x="sepal_length",
    hue="species",
    bins=12,
    edgecolor="gray",
    alpha=0.3,
    element="step",
)

plt.legend()
plt.title("Distribution of Sepal Length for Different Species")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot histograms with different colors for each species
sns.histplot(
    df,
    x="sepal_length",
    hue="species",
    bins=12,
    edgecolor="gray",
    alpha=0.3,
    element="poly",
)

plt.title("Distribution of Sepal Length for Different Species")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

## Adding distribution to histogram

```python
# Plot histograms with different colors for each species
sns.histplot(
    df,
    x="sepal_length",
    hue="species",
    bins=12,
    edgecolor="gray",
    alpha=0.3,
    # fill=False,
    kde=True,
)

plt.title("Distribution of Sepal Length for Different Species")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

# PLotting multiple axis on the same figure

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt


fig, axs = plt.subplots(2, 2, figsize=(7, 7))

sns.histplot(data=df, x="sepal_length", kde=True, color="skyblue", ax=axs[0, 0])
sns.histplot(data=df, x="sepal_width", kde=True, color="olive", ax=axs[0, 1])
sns.histplot(data=df, x="petal_length", kde=True, color="gold", ax=axs[1, 0])
sns.histplot(data=df, x="petal_width", kde=True, color="teal", ax=axs[1, 1])
plt.show()
```
