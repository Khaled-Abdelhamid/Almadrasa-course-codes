```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a DataFrame
data = {
    "Product": ["iPhone", "Samsung Galaxy", "Google Pixel", "OnePlus"],
    "Sales": [4500, 3000, 2000, 500],
}
df = pd.DataFrame(data)
df
```

# Basic pie chart

```python
# df = df.set_index("Product")
df["Sales"].plot(kind="pie")
plt.show()
```

# Pie chart with custom properties

```python
# Example 6: Pie chart with explode (custom wedge properties)
explode = (0.1,0,0,0)  # Pandas doesn't have a direct explode parameter, so we modify matplotlib
colors = ["#4F6272", "#B7C3F3", "#DD7596", "#8EB897"]

fig, ax = plt.subplots()
df.set_index("Product")["Sales"].plot(
    kind="pie",
    labeldistance=1.15,
    wedgeprops={"linewidth": 1, "edgecolor": "white"},
    colors=colors,
    explode=explode,
    autopct="%1.1f%%",  # Display the percentage as floating point number with one decimal place
)
plt.legend(loc="upper left", bbox_to_anchor=(-0.5, 1))  # Move the legend to the left
plt.show()
```
