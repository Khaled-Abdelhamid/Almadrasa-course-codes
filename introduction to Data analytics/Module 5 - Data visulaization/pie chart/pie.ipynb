```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# create data: an array of sales values
sales_values = [1200, 1100, 300, 3000]
```

# Simple pie plot

```python
# Create a pieplot
plt.pie(sales_values)
plt.show()
```

# making custom pie plot

```python
sales_values = [4500, 3000, 2000, 500]
product_names = ["iPhone", "Samsung Galaxy", "Google Pixel", "OnePlus"]
colors = ["#4F6272", "#B7C3F3", "#DD7596", "#8EB897"]

# # Define the explode sequence
# explode = [0.1, 0, 0.3, 0]  # Explode only the first slice (index 0)

# Create the pie chart with the explode parameter and autopct for percentages
plt.pie(
    sales_values,
    labels=product_names,
    labeldistance=1.1,
    # wedgeprops={"linewidth": 1, "edgecolor": "white"},
    colors=colors,
    # explode=explode,
    autopct="%.2f%%",  # Display the percentage as floating point number with one decimal place
)
plt.legend(loc="upper left", bbox_to_anchor=(-0.5, 1))  # Move the legend to the left
plt.show()
```

```python
sales_values = [4500, 3000, 2000, 500]
product_names = ["iPhone", "Samsung Galaxy", "Google Pixel", "OnePlus"]
df =pd.DataFrame(sales_values, index=product_names, columns=["Sales"])
df
```

```python

colors = ["#4F6272", "#B7C3F3", "#DD7596", "#8EB897"]

# # Define the explode sequence
# explode = [0.1, 0, 0.3, 0]  # Explode only the first slice (index 0)

# Create the pie chart with the explode parameter and autopct for percentages
plt.pie(
    x=df["Sales"],
    labels=df.index,
    labeldistance=1.1,
    # wedgeprops={"linewidth": 1, "edgecolor": "white"},
    colors=colors,
    # explode=explode,
    autopct="%.2f%%",  # Display the percentage as floating point number with one decimal place
)
plt.legend(loc="upper left", bbox_to_anchor=(-0.5, 1))  # Move the legend to the left
plt.show()
```
