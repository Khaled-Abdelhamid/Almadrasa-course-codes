```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")
```

 # Encoding numerical columns using point size

```python
# Scatter plot with size by petal width
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="sepal_width",
    size="petal_width",
    sizes=(20, 200),
    alpha=0.7,
)

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset: Sepal Length vs Sepal Width")

# Display the plot
plt.show()
```

```python
# Scatter plot with color by petal length
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="petal_width",
    palette="viridis",
    data=iris,
)

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset: Sepal Length vs Sepal Width")

# Display the plot
plt.show()
```

```python
# Scatter plot with size by petal width and color by petal length
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    size="petal_width",
    sizes=(30, 200),
    hue="petal_width",
    palette="viridis",
    alpha=0.7,
    data=iris,
)

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset: Sepal Length vs Sepal Width")

# Display the plot
plt.show()
```

 # Encoding categorical variables using shapes and colors

```python
# Define a dictionary to map species to markers
markers = {"setosa": "o", "versicolor": "s", "virginica": "D"}

# Scatter plot with customizations
plt.figure(figsize=(10, 6))

sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette=["r", "g", "b"],
    style="species",
    markers=markers,
    data=iris,
    alpha=0.3
)

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset: Sepal Length vs Sepal Width (by Species)")
plt.legend(title="Species")

# Display the plot
plt.show()
```
