```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset("iris")
iris
```

 # Encoding numerical columns using point size

```python
# Scatter plot with customizations
plt.figure(figsize=(10, 6))
sc = plt.scatter(
    iris["sepal_length"],
    iris["sepal_width"],
    s=iris["petal_width"] * 30, 
    alpha=0.7, 
)

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset: Sepal Length vs Sepal Width")

# Display the plot
plt.show()
```

```python
# Scatter plot with customizations
plt.figure(figsize=(10, 6))
sc = plt.scatter(
    iris["sepal_length"],
    iris["sepal_width"],
    c=iris["petal_length"],  # Color by petal length
    cmap="copper",  # Color map
)

# Adding color bar
plt.colorbar(sc, label="Petal Length")

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset: Sepal Length vs Sepal Width")

# Display the plot
plt.show()
```

```python
# Scatter plot with customizations
plt.figure(figsize=(10, 6))
sc = plt.scatter(
    iris["sepal_length"],
    iris["sepal_width"],
    c=iris["petal_length"],  # Color by petal length
    s=iris["petal_length"] * 30,  # Size by petal width
    cmap="viridis",  # Color map
    alpha=0.7,  # Transparency
)

# Adding color bar
plt.colorbar(sc, label="Petal Length")

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset: Sepal Length vs Sepal Width")

# Display the plot
plt.show()
```

 # Encoding categorical variables using shapes and colors

```python
# Define a dictionary to map species to shapes
shapes = {"setosa": "o", "versicolor": "s", "virginica": "D"}
color_map = {"setosa": "r", "versicolor": "g", "virginica": "b"}

# Scatter plot with customizations
plt.figure(figsize=(10, 6))
species = iris["species"].unique()
for species in species:
    subset = iris[iris["species"] == species]
    plt.scatter(
        subset["sepal_length"],
        subset["sepal_width"],
        c=color_map[species],
        marker=shapes[species],
        label=species,
        alpha=0.5,  # Transparency
    )

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset: Sepal Length vs Sepal Width (by Species)")
plt.legend(title="Species")

# Display the plot
plt.show()
```
