```python
# libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df
```

# Simple boxplot

```python
df["sepal_length"].plot.box(vert=True)
plt.ylabel("Sepal Length")
plt.title("Boxplot of Sepal Length")
plt.show()
```

# Boxplot by species

```python
df.boxplot(column="sepal_length", by="species")
plt.ylabel("Sepal Length")
plt.title("Boxplot of Sepal Length by Species")
plt.suptitle("")
plt.show()
```

# Custom Box plot

```python
# Assuming df is your DataFrame
virginica_subset = df[df["species"] == "virginica"]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create a customized boxplot using pandas
virginica_subset["sepal_length"].plot(
    kind="box",
    ax=ax,
    color={"boxes": "lightblue", "whiskers": "blue", "medians": "red", "caps": "gold"},
    flierprops={"marker": "D", "markerfacecolor": "red", "markersize": 1},
    meanline=True,
    showmeans=True,
)

# Set labels and title
ax.set_xlabel("Column")
ax.set_ylabel("Sepal Length")
ax.set_title("Boxplot of Virginica Sepal Length")

# Show the plot
plt.show()
```
