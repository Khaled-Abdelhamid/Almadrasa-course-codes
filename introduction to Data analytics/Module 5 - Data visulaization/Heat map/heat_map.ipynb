```python
# library
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Create a dataset
df = pd.DataFrame(np.random.random((5, 5)), columns=["a", "b", "c", "d", "e"])
df
```

# Simple Heatmap

```python
# Default heatmap
# sns.heatmap(df)
# sns.heatmap(df, annot=True)
sns.heatmap(df, annot=True, fmt=".4f")
```

# Change annotation font size

```python
# plot a heatmap with annotation
sns.heatmap(df, annot=True, annot_kws={"size":20})
plt.show()
```

# Change line color

```python
sns.heatmap(df, linewidths=2, linecolor="yellow")
plt.show()
```

# Visualizing correlation using heatmap

```python
corr_matrix = df.corr()
corr_matrix
```

```python
# plot it
sns.heatmap(data=corr_matrix, annot=True)
plt.show()
```

## Use diverging colors with correlation

```python
# plot it
sns.heatmap(corr_matrix,cmap="coolwarm", annot=True)
plt.show()
```

# Changing color scale

```python
# Create a dataset
df = pd.DataFrame(np.random.random((5, 5)), columns=["a", "b", "c", "d", "e"])
df
```

```python
fig, axs = plt.subplots(
    nrows=1, ncols=3,
    figsize=(12, 3)
)

sns.heatmap(df, vmin=0, vmax=1, annot=True ,ax=axs[0])
axs[0].set_title('Default heatmap with range 0 to 1')

sns.heatmap(df, vmin=0.5, vmax=1, annot=True ,ax=axs[1])
axs[1].set_title('Dark heatmap with range 0.5 to 1')

sns.heatmap(df, vmin=0, vmax=0.5, annot=True ,ax=axs[2])
axs[2].set_title('Light heatmap with range 0 to 0.5')

plt.show()
```

# Heat map on  pivoting tables

```python
df = pd.DataFrame(
    {
        "feature": list(range(1, 6)) * 5,
        "people": np.repeat(("A", "B", "C", "D", "E"), 5),
        "value": np.random.random(25),
    }
)
df
```

```python
# Turn long format into a wide format
df_pivoted = df.pivot_table(index="feature", columns="people", values="value")
df_pivoted
```

```python
# plot it
sns.heatmap(df_pivoted,annot=True)
plt.show()
```
