```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# Simple plot with one dimensional data

```python
values = np.cumsum(np.random.randn(1000))
```

```python
# use seaborn's lineplot function
sns.lineplot(data=values)

# show the graph
plt.show()
```

# line plot with x and y value

```python
df = pd.DataFrame({"xvalues": range(1, 101), "yvalues": np.random.randn(100)})
```

```python
# plot
# sns.lineplot(x="xvalues", y="yvalues", data=df)
sns.lineplot(x=df["xvalues"], y=df["yvalues"])

# show the graph
plt.show()
```

 # Use line plot with only ordered data on the x axis

```python
# import the iris dataset
df = sns.load_dataset("iris")
df
```

```python

sns.scatterplot(x="sepal_width", y="sepal_length", data=df,color = "red",alpha = 0.3)

# plot
sns.lineplot(x="sepal_width", y="sepal_length", data=df)

# show the graph
plt.show()
```
