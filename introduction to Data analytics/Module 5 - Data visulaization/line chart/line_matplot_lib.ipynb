# Time series Data

```python
# Init an empty list that will stores the dates
dates = []

# Iterates over our range and add the value to the list
for date in range(1970, 2023):
    dates.append(str(date))
dates
# Generate a random variable
sample_size = 2023 - 1970
variable = np.random.normal(
    100,  # mean
    15,  # standard deviation
    sample_size,  # sample size
)
# variable
df = pd.DataFrame(data={"date": dates, "value": variable})
df
```

```python
# Set the fig size
plt.figure(figsize=(20, 5))

# Create the line chart
plt.plot(df["date"], df["value"])

# Add labels and a title
plt.xlabel(xlabel="Date")
plt.ylabel(ylabel="Sales")
plt.title(label="Basic Line Chart")

# Rotate x-axis labels by 30 degrees for better visibility
plt.xticks(rotation=30)

# Add a text annotation
plt.text(
    x=df["date"].iloc[38],  # x-axis position
    y=df["value"].iloc[38]+10,  # y-axis position
    s="What a nice chart!",  # text displayed
    fontsize=13,
    color="red",
)

# # Find the index of the highest value
highest_index = df["value"].idxmax()

# # Add a circle annotation at the highest value
plt.scatter(
    df["date"].iloc[highest_index],
    df["value"].iloc[highest_index],
    color="red",
    marker="o",  # Specify that we want a circle
    s=100,  # Size
    label = "Highest Value"  # Label
)

# # Calculate the median value
median_value = df["value"].median()

# # Add a reference line at the median value
plt.axhline(
    y=median_value, color="green", linestyle="--", label="Reference Line (Median)"
)
plt.legend()  # Add a legend
# # Display the chart
plt.show()
```

# Fill Area between graphs 

```python
data = {
    "time": range(12),
    "income": [5, 9, 6, 6, 10, 7, 6, 4, 4, 5, 6, 4],
    "expenses": [6, 6, 8, 3, 6, 9, 7, 8, 6, 6, 4, 8],
}
df = pd.DataFrame(data)
df.head()
```

```python
# Initialize figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Plot lines
ax.plot(df["time"], df["income"], color="black", label="Income")
ax.plot(df["time"], df["expenses"], color="yellow", label="Expenses")

# Fill area when income > expenses with green
ax.fill_between(
    df["time"],
    df["income"],
    df["expenses"],
    where=(df["income"] > df["expenses"]),
    interpolate=True,
    color="green",
    alpha=0.25,
    label="Positive",
)

# Fill area when income <= df['expenses'] with red
ax.fill_between(
    df["time"],
    df["income"],
    df["expenses"],
    where=(df["income"] <= df["expenses"]),
    interpolate=True,
    color="red",
    alpha=0.25,
    label="Negative",
)

ax.legend()
plt.show()
```
