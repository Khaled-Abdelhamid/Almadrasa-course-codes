In Matplotlib, a plotting library in Python, both `Figure` and `Axes` play crucial roles in the creation of plots, but they serve different purposes and operate at different levels of abstraction.

### Figure

A `Figure` is essentially the whole window or page that everything else is drawn on. It contains all the elements of the plot such as labels, legends, title, and the entire set of axes. A `Figure` can contain multiple `Axes` objects, allowing for complex layouts with multiple subplots. Each `Figure` is independent and can exist without any associated `Axes`. However, once `Axes` are added to a `Figure`, they become part of that `Figure`.

Key points about `Figure`:
- Acts as a container for all plot elements.
- Can contain multiple `Axes`.
- Independent entity before `Axes` are added.
- Represents the whole window/page in a GUI application.

### Axes

An `Axes` is where the actual plotting happens. It is responsible for drawing the plot itself, including the plot area, labels, tick marks, and grid. In other words, an `Axes` is a single plot or graph within a `Figure`. When you create a simple plot, you are actually creating a single `Axes` object. If you want to create a subplot or a grid of plots, you would add multiple `Axes` objects to your `Figure`.

Key points about `Axes`:
- Responsible for the actual plotting area.
- Can be thought of as a single plot or graph.
- Can be added to a `Figure` to create complex layouts.
- Multiple `Axes` can coexist within a single `Figure`.

### Summary

- **Figure**: The top-level container for all plot elements. It represents the entire window or page in a graphical environment. A `Figure` can contain multiple `Axes`.
- **Axes**: The area where the actual plotting occurs. An `Axes` is a single plot or graph within a `Figure`. Multiple `Axes` can be added to a `Figure` to create complex layouts.

This distinction allows for great flexibility in creating plots, from simple graphs to complex multi-panel figures with shared scales, colorbars, and more.

```python
# Importing the matplotlib.pyplot module, which provides a MATLAB-like interface for making plots and charts.
import matplotlib.pyplot as plt

# Creating a figure object and an axes object using plt.subplots().
# plt.subplots() returns a tuple containing a Figure object and one or more Axes objects.
fig, ax = plt.subplots()

# Printing the type of the figure object to confirm it's a matplotlib.figure.Figure instance.
print(type(fig))  # Should output: <class 'matplotlib.figure.Figure'>

# Printing the type of the axes object to confirm it's a matplotlib.axes.Axes instance.
print(type(ax))  # Should output: <class 'matplotlib.axes.Axes'>
```

```python
# 2019 data
df_19 = pd.read_csv('2019.csv')
df_19.columns = ['Rank','Country','Score','GDP',
                'Support','Health',
                'Freedom','Generosity','Corruption']

df_19.head()
```

```python
# Data processing
top_economies = ['United States', 'China','Japan', 'Germany','United Kingdom','India', 'France','Brazil', 'Canada']
df_top = df_19[(df_19['Country'].isin(top_economies))].sort_values(['Country'])
df_top.reset_index(drop=True)

x = df_top.loc[:, ['Score']]
df_top['Score_z'] = (x - x.mean())/x.std()
df_top['colors'] = ['red' if x < 0 else 'blue' for x in df_top['Score_z']]
df_top.sort_values('Score_z', inplace=True)
df_top.reset_index(inplace=True)

# Draw plot
plt.figure(figsize=(8,6), dpi= 50)
plt.hlines(y=df_top.index, xmin=0, xmax=df_top.Score_z, color=df_top.colors, alpha=0.5, linewidth=20)

# Decorations
plt.gca().set(ylabel='Country', xlabel='Happiness_Score')
plt.yticks(df_top.index, df_top.Country, fontsize=15)
plt.title('Happiness Score for Top Economies in 2019', fontdict={'size':20})
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```
