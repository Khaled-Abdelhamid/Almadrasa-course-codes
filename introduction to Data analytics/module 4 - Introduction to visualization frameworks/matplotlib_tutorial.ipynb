- **Figure**: 
  - Acts as a container for all plot elements and represents the entire window/page.
  - Can contain multiple `Axes`, allowing for complex layouts.

- **Axes**: 
  - Responsible for the actual plotting area, including labels, tick marks, and grids.
  - Can be added to a `Figure` to create complex layouts with multiple plots or graphs.

```python
# Creating a figure object and an axes object using plt.subplots().
# plt.subplots() returns a tuple containing a Figure object and one or more Axes objects.
fig, ax = plt.subplots(figsize=(20, 2))

# Printing the type of the figure object to confirm it's a matplotlib.figure.Figure instance.
print(type(fig))  # Should output: <class 'matplotlib.figure.Figure'>

# Printing the type of the axes object to confirm it's a matplotlib.axes.Axes instance.
print(type(ax))  # Should output: <class 'matplotlib.axes.Axes'>
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.linspace(0.1, 10, 1000)

# Define y-axis formulas as variables
y_linear = 2 * x
y_quadratic = x + 3
y_logarithmic = np.log(x)
y_sine = np.sin(x)
y_cosine = np.cos(x)

# Plot with different colors and linestyles
plt.plot(
    x, y_linear, color="blue", linestyle="--"
)  # specify color by name, dashed style
plt.plot(
    x, y_quadratic, color="g", linestyle=":"
)  # short color code (rgbcmyk), dotted style
plt.plot(
    x, y_logarithmic, color="0.75", linestyle="-."
)  # Grayscale between 0 and 1, dashdot style
plt.plot(
    x, y_sine, color="#FFDD44", linestyle="--"
)  # Hex code (RRGGBB from 00 to FF), dashed style

# Display the plot
plt.show()
```

 In this example, we plot the sine and cosine functions on the same plot. We add labels to the x-axis and y-axis, set a title for the plot, and include a legend to describe the plotted lines. The legend helps in identifying which line corresponds to which mathematical function.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.linspace(0.1, 10, 1000)

# Define y-axis formulas as variables
y_linear = 2 * x
y_quadratic = x + 3
y_logarithmic = np.log(x)
y_sine = np.sin(x)
y_cosine = np.cos(x)

# Plot with different colors and linestyles
plt.plot(
    x, y_linear, color="blue", linestyle="--"
)  # specify color by name, dashed style
plt.plot(
    x, y_quadratic, color="g", linestyle=":"
)  # short color code (rgbcmyk), dotted style
plt.plot(
    x, y_logarithmic, color="0.75", linestyle="-."
)  # Grayscale between 0 and 1, dashdot style
plt.plot(
    x, y_sine, color="#FFDD44", linestyle="--"
)  # Hex code (RRGGBB from 00 to FF), dashed style

# Adding annotations
# Example 1: Simple text annotation
plt.annotate(
    xy=(x.max(), y_linear.max()),
    text="Maximum Value",
    xytext=(x.max() + 1, y_linear.max()),
    fontsize=12,
    fontweight="bold",
    arrowprops={"facecolor": "black", "edgecolor": "red"},
)

# Example 2: Annotation with custom text properties
plt.annotate(
    xy=(x.min(), y_linear.min()),
    text="Minimum Value",
    xytext=(x.min() - 1, y_linear.min() - 2),
    fontsize=12,
    fontweight="bold",
    arrowprops={"facecolor": "red", "edgecolor": "black"},
)
# Display the plot
plt.show()
```

 ## Creating Subplots and Facet Grids



 Subplots and facet grids are useful for creating multiple plots in a single figure. This section demonstrates how to create and manage multiple plots using `plt.subplots()` and Seaborn's `FacetGrid`.

```python

# Create subplots with a 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs.shape
# Generate random data for plotting
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log(x + 1)

# Plot different data on each subplot
axs[0, 0].plot(x, y1, "r")
axs[0, 0].set_title("Sine")

axs[0, 1].plot(x, y2, "b")
axs[0, 1].set_title("Cosine")

axs[1, 0].plot(x, y3, "g")
axs[1, 0].set_title("Tangent")

axs[1, 1].plot(x, y4, "k")
axs[1, 1].set_title("Logarithm")

# Display the plot
plt.tight_layout()
plt.show()
```

 In this example, we create a figure with a 2x2 grid of subplots using `plt.subplots()`. We generate different mathematical functions and plot each on a different subplot within the grid. The `plt.tight_layout()` function ensures that the subplots are neatly arranged without overlapping.
