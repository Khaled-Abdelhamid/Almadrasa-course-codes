 # Seaborn Tutorial
 This tutorial will cover the basics of using Seaborn, a powerful Python library for data visualization built on top of Matplotlib. We will go through the setup, basic plots, advanced plots, and customization options.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
```

 ## Introduction to Seaborn



 Seaborn is a Python visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn is well-suited for data analysis and visualization, offering a range of plots and styles.

 ## Loading Data



 Let's start by loading some example data. Seaborn comes with several built-in datasets.

```python
# Load an example dataset from Seaborn
tips = sns.load_dataset("tips")
tips.head()
```

 ## Basic Plots



 ### Scatter Plot



 Scatter plots are used to display the relationship between two variables.

```python
# Create a scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")
ax=plt.gca()
print(type(ax))
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

 ### Line Plot



 Line plots are useful for visualizing data points connected by straight lines.

```python
# Create a line plot
sns.lineplot(data=tips, x="size", y="total_bill")
plt.title("Line Plot of Total Bill vs Tip")
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size as needed
plt.plot(tips["size"], tips["total_bill"])
plt.title("Line Plot of Total Bill vs Tip")
plt.xlabel("Size")
plt.ylabel("Total Bill")
plt.grid(True)
plt.show()
```

 ## Advanced Plots



 ### Pair Plot



 Pair plots are useful for visualizing the relationship between multiple variables.

```python
# Create a pair plot
sns.pairplot(tips)
plt.suptitle("Pair Plot of Tips Dataset")
plt.show()
```

 ### Facet Grid



 Facet grids allow you to create multiple subplots based on the values of one or more categorical variables.

```python
# Create a facet grid
g = sns.FacetGrid(tips, col="sex", row="time")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

 ## Customization



 Seaborn allows for extensive customization of plots, including changing colors, styles, and adding annotations.

 ### Color Palette



 You can customize the color palette of your plots.

```python
# Create a bar plot instead of a scatter plot for better visibility
sns.barplot(x="size", y="tip", data=tips, palette="deep")
plt.title("Bar Plot with Custom Color Palette")
plt.show()
```

 ### Adding Annotations



 You can add annotations to your plots to highlight important points or provide additional information.

```python

# Add annotations to a plot
plot = sns.scatterplot(data=tips, x="total_bill", y="tip")
for line in range(0, tips.shape[0]):
    plot.text(
        tips.total_bill[line] + 0.2,
        tips.tip[line],
        tips.index[line],
        horizontalalignment="left",
        size="small",
        color="black",
        weight="semibold",
    )
plt.title("Scatter Plot with Annotations")
plt.show()
```
